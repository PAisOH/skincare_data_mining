### 1.4.1 根据正则切分about
- 比较顺利，about比较规整

```python
# 还是有问题“Dream Lip Oil for Moisturizing Sheer Coverage”
df = pd.read_csv('../Desktop/DBCosmetic/Sephora_FINAL.csv')

# 定义一个函数来提取每个字段的内容
def extract_field(about, field, alt_field=None):
    # 检查about是否为字符串
    if not isinstance(about, str):
        return None
    
    # 尝试查找主字段
    start = about.find(field + ':')
    if start != -1:
        start += len(field) + 1
        end = about.find('\n\n', start)
        if end == -1:
            return about[start:].strip()
        return about[start:end].strip()
    
    # 如果主字段不存在,尝试查找备用字段
    if alt_field:
        start = about.find(alt_field + ':')
        if start != -1:
            start += len(alt_field) + 1
            end = about.find('\n\n', start)
            if end == -1:
                return about[start:].strip()
            return about[start:end].strip()
    
    # 如果两个字段都不存在,返回空字符串
    return None

# 创建新字段
df['What it is'] = df['about'].apply(lambda x: extract_field(x, 'What it is'))
df['Skin Type'] = df['about'].apply(lambda x: extract_field(x, 'Skin Type'))
df['Skincare Concerns'] = df['about'].apply(lambda x: extract_field(x, 'Skincare Concerns'))
df['Formulation'] = df['about'].apply(lambda x: extract_field(x, 'Formulation', 'Formulation Type'))
df['Benefits'] = df['about'].apply(lambda x: extract_field(x, 'Benefits'))
df['Highlighted Ingredients'] = df['about'].apply(lambda x: extract_field(x, 'Highlighted Ingredients'))
df['Ingredient Callouts'] = df['about'].apply(lambda x: extract_field(x, 'Ingredient Callouts'))
df['What Else You Need to Know'] = df['about'].apply(lambda x: extract_field(x, 'What Else You Need to Know'))
df['Clinical Results'] = df['about'].apply(lambda x: extract_field(x, 'Clinical Results'))

df.to_csv('../Desktop/DBCosmetic/Sephora_Proccessing.csv', index=False)
```

```python
def process_strings(input_strings):
    processed_strings = []
    for input_string in input_strings:
        # 1. 所有英文全部改成小写字母
        lowercase_string = input_string.lower()
        
        # 2. 找到所有ci 加上一串数字的列名
        import re
        ci_pattern = r'ci\s(\d+)'
        ci_match = re.search(ci_pattern, lowercase_string)
        if ci_match:
            ci_number = ci_match.group(1)
            processed_string = f'ci {ci_number}'
        else:
            processed_string = lowercase_string
        
        # 3. 去掉所有首尾空格
        processed_string = processed_string.strip()
        
        # 4. 去掉所有括号及括号内的内容
        processed_string = re.sub(r'\([^)]*\)', '', processed_string)
        
        # 5. 所有空格改成"-"连接
        processed_string = processed_string.replace(' ', '-')
        
        processed_strings.append(processed_string)
    
    return processed_strings

# 测试用例
input_strings = ["Ci 15985(yellow 6).", "Vas(Ci 19721)", "Gum ball"]
output_strings = process_strings(input_strings)
print(output_strings)  # ['ci-15985', 'ci-19721', 'gum-ball']
```

```python
# TEST

import pandas as pd
from fuzzywuzzy import fuzz
from tqdm import tqdm
import re

# 读取数据
df_a = pd.read_csv('../Desktop/DBCosmetic/Sephora_Proccessing_updated.csv').head(3)
df_b = pd.read_csv('../Desktop/Paula_s_Choice/Paula_SUM_LIST.csv')

# 复制一份a表,叫做df_a_all
df_a_all = df_a.copy()
df_a_all['new_ingredients'] = ''  # 初始化new_ingredients列

# 遍历df_a_all每一行的clean_ingredients
for index, row in tqdm(df_a_all.iterrows(), total=len(df_a_all), desc='Processing df_a_all'):
    if pd.notnull(row['clean_ingredients']):  # 检查clean_ingredients是否为空
        ingredients = str(row['clean_ingredients'])
        
        # 1. 先把所有括号和括号内的内容删除
        ingredients = re.sub(r'\([^)]*\)', '', ingredients)
        
        ingredient_list = ingredients.split(', ')
        
        new_ingredients = []
        remaining_ingredients = []
        for ingredient in ingredient_list:
            # 以单词为最小单位进行模糊匹配
            ingredient_words = ingredient.lower().split()
            match_found = False
            for b_ingredient in df_b['ingredient_name']:
                b_ingredient_words = str(b_ingredient).lower().split()
                if all(word in b_ingredient_words for word in ingredient_words):
                    new_ingredients.append(b_ingredient)
                    match_found = True
                    break
            if not match_found:
                # 使用 fuzzywuzzy 进行模糊匹配
                highest_match_score = 0
                highest_match_ingredient = None
                for b_ingredient in df_b['ingredient_name']:
                    match_score = fuzz.token_set_ratio(' '.join(ingredient_words), str(b_ingredient).lower())
                    if match_score > highest_match_score:
                        highest_match_score = match_score
                        highest_match_ingredient = b_ingredient
                if highest_match_score >= 80:
                    new_ingredients.append(highest_match_ingredient)
                else:
                    remaining_ingredients.append(ingredient)
        df_a_all.at[index, 'new_ingredients'] = ', '.join(new_ingredients)
        df_a_all.at[index, 'clean_ingredients'] = ', '.join(remaining_ingredients)

# 复制一份a表,叫做df_a_link
df_a_link = df_a.copy()
df_a_link['new_ingredients'] = ''  # 初始化new_ingredients列

# 遍历df_a_link每一行的clean_ingredients
for index, row in tqdm(df_a_link.iterrows(), total=len(df_a_link), desc='Processing df_a_link'):
    if pd.notnull(row['clean_ingredients']):
        ingredients = str(row['clean_ingredients'])
        
        # 1. 先把所有括号和括号内的内容删除
        ingredients = re.sub(r'\([^)]*\)', '', ingredients)
        
        ingredient_list = ingredients.split(', ')
        
        new_ingredients = []
        remaining_ingredients = []
        for ingredient in ingredient_list:
            # 以单词为最小单位进行模糊匹配
            ingredient_words = ingredient.lower().split()
            match_found = False
            for b_ingredient in df_b[df_b['link'].notnull()]['ingredient_name']:
                b_ingredient_words = str(b_ingredient).lower().split()
                if all(word in b_ingredient_words for word in ingredient_words):
                    new_ingredients.append(b_ingredient)
                    match_found = True
                    break
            if not match_found:
                # 使用 fuzzywuzzy 进行模糊匹配
                highest_match_score = 0
                highest_match_ingredient = None
                for b_ingredient in df_b[df_b['link'].notnull()]['ingredient_name']:
                    match_score = fuzz.token_set_ratio(' '.join(ingredient_words), str(b_ingredient).lower())
                    if match_score > highest_match_score:
                        highest_match_score = match_score
                        highest_match_ingredient = b_ingredient
                if highest_match_score >= 80:
                    new_ingredients.append(highest_match_ingredient)
                else:
                    remaining_ingredients.append(ingredient)
        df_a_link.at[index, 'new_ingredients'] = ', '.join(new_ingredients)
        df_a_link.at[index, 'clean_ingredients'] = ', '.join(remaining_ingredients)

# 保存结果
df_a_all.to_csv('../Desktop/DBCosmetic/Sephora_all.csv', index=False)
df_a_link.to_csv('../Desktop/DBCosmetic/Sephora_linKNOTNull.csv', index=False)
```

```python
import pandas as pd
from fuzzywuzzy import fuzz
from tqdm import tqdm
import re
import multiprocess as mp

# 读取数据
df_a = pd.read_csv('../Desktop/DBCosmetic/Sephora_Proccessing_updated.csv')
df_b = pd.read_csv('../Desktop/Paula_s_Choice/Paula_embedding_SUMLIST_before_422.csv')

# 定义一个函数来处理单个dataframe
def process_dataframe(df):
    df['new_ingredients'] = ''  # 初始化new_ingredients列
    
    for index, row in tqdm(df.iterrows(), total=len(df), desc=f'Processing {mp.current_process().name}'):
        if pd.notnull(row['clean_ingredients']):  # 检查clean_ingredients是否为空
            ingredients = str(row['clean_ingredients'])
            
            # 1. 先把所有括号和括号内的内容删除,同时保存原始列表
            original_ingredients = ingredients.split(', ')
            ingredients = re.sub(r'\([^)]*\)', '', ingredients)
            ingredient_list = ingredients.split(', ')
            
            matched_indices = []
            new_ingredients = []
            remaining_ingredients = []
            for i, ingredient in enumerate(ingredient_list):
                # 以单词为最小单位进行模糊匹配
                ingredient_words = ingredient.lower().split()
                match_found = False
                for b_ingredient in df_b['ingredient_name']:
                    b_ingredient_words = str(b_ingredient).lower().split()
                    if all(word in b_ingredient_words for word in ingredient_words):
                        new_ingredients.append(b_ingredient)
                        matched_indices.append(i)
                        match_found = True
                        break
                if not match_found:
                    # 使用 fuzzywuzzy 进行模糊匹配
                    highest_match_score = 0
                    highest_match_ingredient = None
                    for b_ingredient in df_b['ingredient_name']:
                        match_score = fuzz.token_set_ratio(' '.join(ingredient_words), str(b_ingredient).lower())
                        if match_score > highest_match_score:
                            highest_match_score = match_score
                            highest_match_ingredient = b_ingredient
                    if highest_match_score >= 80:
                        new_ingredients.append(highest_match_ingredient)
                        matched_indices.append(i)
                    else:
                        remaining_ingredients.append(original_ingredients[i])
            
            # 重建clean_ingredients
            df.at[index, 'new_ingredients'] = ', '.join(new_ingredients)
            df.at[index, 'clean_ingredients'] = ', '.join(remaining_ingredients)
    return df

# 创建进程池并处理数据
num_processes = mp.cpu_count() * 2
pool = mp.Pool(processes=num_processes)
df_a_all = pool.apply_async(process_dataframe, args=(df_a.copy(),)).get()
df_a_link = pool.apply_async(process_dataframe, args=(df_a.copy(),)).get()
pool.close()
pool.join()

# 保存结果
df_a_all.to_csv('../Desktop/DBCosmetic/Sephora_all_423.csv', index=False)
df_a_link.to_csv('../Desktop/DBCosmetic/Sephora_linKNOTNull_423.csv', index=False)
```

```python
# 拆分成Neo4J方便处理的link表
import pandas as pd

# 读取CSV文件
df = pd.read_csv('../Desktop/DBCosmetic/Sephora_all_425.csv')

# 去除'new_ingredients'为空的数据行
df = df[~df['new_ingredients'].isna()]

# 创建目标表格
rows = []

# 先处理包含','的成分
for index, row in df.iterrows():
    cosmetic_name = row['cosmetic_name']
    new_ingredients = row['new_ingredients']
    
    # 处理'Citrus Paradisi, X C. Reticulata Peel Oil Expressed'
    if 'Citrus Paradisi, X C. Reticulata Peel Oil Expressed' in new_ingredients:
        new_ingredients = new_ingredients.replace('Citrus Paradisi, X C. Reticulata Peel Oil Expressed', 'CCCCCCCCCC1')
    # 处理'1, 2-Hexanediol'
    if '1, 2-Hexanediol' in new_ingredients:
        new_ingredients = new_ingredients.replace('1, 2-Hexanediol', 'CCCCCCCCCC2')
    
    # 切分其他成分并加入新表格
    for ingredient in new_ingredients.split(', '):
        rows.append({'cosmetic': cosmetic_name, 'ingredient': ingredient.strip()})

target_df = pd.DataFrame(rows)

# 将特殊标志替换回真实成分名
target_df['ingredient'] = target_df['ingredient'].replace('CCCCCCCCCC1', 'Citrus Paradisi, X C. Reticulata Peel Oil Expressed')
target_df['ingredient'] = target_df['ingredient'].replace('CCCCCCCCCC2', '1, 2-Hexanediol')

# 检查重复数据
duplicate_rows = target_df.duplicated(subset=['cosmetic', 'ingredient'], keep='first')
target_df = target_df[~duplicate_rows]
print(f"Removed {duplicate_rows.sum()} duplicate rows.")

# 保存更新后的表格到CSV
target_df.to_csv('../Desktop/DBCosmetic/binary_cosmetic_ingredient.csv', index=False)
```
