## 1. upset图

```python
import pandas as pd
import matplotlib.pyplot as plt
from upsetplot import plot, from_contents

df = pd.read_csv('../Desktop/Paula_s_Choice/Paula_DETAIL_FINAL_419.csv')

# 2. 排除 categories 列为空的行
empty_rows = df[df['categories'].isnull() | (df['categories'] == '')]
print(f"Number of rows with empty 'categories': {len(empty_rows)}")
df = df[~(df['categories'].isnull() | (df['categories'] == ''))]

# 3. 创建数据字典
contents = {}
for index, row in df.iterrows():
    categories = row['categories'].split(', ')
    for category in categories:
        if category not in contents:
            contents[category] = []
        contents[category].append(index)

# 4. 绘制 UpSet 图
plt.figure(figsize=(12, 8))
plot(from_contents(contents), orientation='vertical', sort_by='cardinality')
plt.title("Categories")

# 5. 手动计算前 10 个重合率最高的组合
intersection_counts = {}
for key1, value1 in contents.items():
    for key2, value2 in contents.items():
        if key1 < key2:  # 确保只计算一次
            intersection = set(value1) & set(value2)
            intersection_count = len(intersection)
            if intersection_count > 0:
                intersection_counts[(key1, key2)] = intersection_count

# 6. 输出前 10 个重合率最高的组合
sorted_intersections = sorted(intersection_counts.items(), key=lambda x: x[1], reverse=True)
print("Top 10 intersections:")
for i in range(min(10, len(sorted_intersections))):
    print(f"{sorted_intersections[i][0]}: {sorted_intersections[i][1]}")

# 7. 保存图片
plt.savefig('../desktop/upset_plot2.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 导入表格数据
df = pd.read_csv('../Desktop/Paula_s_Choice/Paula_DETAIL_FINAL_419.csv')

# 2. 排除 categories 列为空的行
empty_rows = df[df['categories'].isnull() | (df['categories'] == '')]
print(f"Number of rows with empty 'categories': {len(empty_rows)}")
df = df[~(df['categories'].isnull() | (df['categories'] == ''))]

# 3. 创建分类计数矩阵
category_list = []
for index, row in df.iterrows():
    categories = row['categories'].split(', ')
    category_list.extend(categories)
unique_categories = pd.unique(category_list)

category_counts = pd.DataFrame(0, index=unique_categories, columns=unique_categories)
for index, row in df.iterrows():
    categories = row['categories'].split(', ')
    for i in range(len(categories)):
        for j in range(i+1, len(categories)):
            category_counts.loc[categories[i], categories[j]] += 1
            category_counts.loc[categories[j], categories[i]] += 1

# 4. 绘制热力图
plt.figure(figsize=(16, 14))
sns.heatmap(category_counts, cmap="YlOrRd", annot=True, fmt=".0f", vmin=0, vmax=category_counts.max().max())
plt.title("Heatmap of Category Relationships", fontsize=16)
plt.xlabel("Categories", fontsize=14)
plt.ylabel("Categories", fontsize=14)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)
# plt.savefig('category_heatmap.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import linkage, dendrogram

# 1. 导入表格数据
df = pd.read_csv('../Desktop/Paula_s_Choice/Paula_DETAIL_FINAL_419.csv')

# 2. 排除 categories 列为空的行
empty_rows = df[df['categories'].isnull() | (df['categories'] == '')]
print(f"Number of rows with empty 'categories': {len(empty_rows)}")
df = df[~(df['categories'].isnull() | (df['categories'] == ''))]

# 3. 创建分类计数矩阵
category_list = []
for index, row in df.iterrows():
    categories = row['categories'].split(', ')
    category_list.extend(categories)
unique_categories = pd.unique(category_list)

category_counts = pd.DataFrame(0, index=unique_categories, columns=unique_categories)
for index, row in df.iterrows():
    categories = row['categories'].split(', ')
    for i in range(len(categories)):
        for j in range(i+1, len(categories)):
            category_counts.loc[categories[i], categories[j]] += 1
            category_counts.loc[categories[j], categories[i]] += 1

# 4. 绘制聚类热力图
plt.figure(figsize=(16, 14))

# 计算分类之间的相关性矩阵
corr_matrix = category_counts.corr()

# 计算层次聚类
linkage_matrix = linkage(corr_matrix.values, 'ward')

# 绘制聚类热力图
sns.clustermap(corr_matrix, cmap="YlOrRd", figsize=(16, 14), dendrogram_ratio=.2, cbar_pos=(0, .2, .03, .4))
plt.title("Clustered Heatmap of Category Relationships")
# plt.savefig('category_clustered_heatmap.png', dpi=300)
plt.show()
```

```python
# 进行分割

# 读取数据
final_df = pd.read_csv('../Desktop/Paula_s_Choice/Paula_detail_422.csv')

# 创建一个新的列 'description_new'
final_df['description_new'] = ''

for index, row in final_df.iterrows():
    # 1. 定位到 'Description' 单词所在位置,删除包括这个单词之前的内容
    start = row['all'].find('Description')
    if start != -1:
        start += len('Description')
        row_text = row['all'][start:]
    else:
        row_text = row['all']
    
    # 2. 定位到 'Written by:' 所在位置,删除包括 'Written by:' 之后的内容
    end = row_text.find('Written by:')
    if end != -1:
        row_text = row_text[:end]

    # 3. 删除首尾的空格和换行符
    row_text = row_text.strip()

    # 4. 检查 'description_new' 是否包含自然语言
    if not bool(re.search(r'\w', row_text)):
        row_text = None
    
    # 将处理后的文本赋值给新的列 'description_new'
    final_df.at[index, 'description_new'] = row_text

# 保存修改后的DataFrame到新的CSV文件
final_df.to_csv('../Desktop/Paula_s_Choice/Paula_detail_final_422.csv', index=False)
```
